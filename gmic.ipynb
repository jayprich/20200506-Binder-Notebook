```
from IPython.display import display, Image
from glob import glob
import gmic

!echo nothing > /tmp/o.png

def show_example(script):
    !rm /tmp/o*.png
    gmic.run('sample_swan.png james_prichard.gmic ' + script + ' cut 0,255 o /tmp/o.png')
    display(*[Image(f) for f in glob('/tmp/o*.png')])
```

```
show_example("""
+jpr_phasecongruence. 3,7,30,0
+jpr_phasecongruence.. 60,1,50,0
jpr_phasecongruence... 7,17,130,1
""")
```
